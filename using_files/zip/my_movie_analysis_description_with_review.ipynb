```python
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import PromptTemplate
from langchain_anthropic import ChatAnthropic

# 定义一个函数
def isRelated(movie_reviews,movie_description):
  # 模型
  llm = ChatAnthropic(temperature=0,model_name="claude-3-opus-20240229")
  # 修改之后的prompt模板
  prompt = PromptTemplate.from_template("""there are movie_reviews:
  {movie_reviews},

  and these are movie_description:
  {movie_description}.

  If you think they describe the same movie, please respond with "Y". If not, please respond with "N". No further information is needed.""")
  # chain
  my_chain =  prompt | llm | StrOutputParser()
  return my_chain.invoke({"movie_reviews": movie_reviews,"movie_description": movie_description})
```

```python
import json
data_collection = []
for index_y, row_y in df_reviews.iterrows():
  review = row_y['review']
  review_id = str(row_y['review_id'])
  data = review_id+"#"+review
  data_collection.append(data)
data_collection
```

```python
# https://docs.anthropic.com/claude/docs/embeddings
import voyageai
vo = voyageai.Client()
doc_embds = vo.embed(
    data_collection, model="voyage-2", input_type="document"
).embeddings
doc_embds
```

```python
# 测试输出
import numpy as np

query = "A scrappy, feisty, fiercely loyal Chicago family makes no apologies."

# Embed the query
query_embd = vo.embed(
    query, model="voyage-2", input_type="query"
).embeddings[0]
similarities = np.dot(doc_embds, query_embd)
retrieved_id = np.argmax(similarities)
print(data_collection[retrieved_id])
```

```python
# 写个函数
def getRelateTrue(query):
  # Embed the query
  query_embd = vo.embed(
      query, model="voyage-2", input_type="query"
  ).embeddings[0]
  similarities = np.dot(doc_embds, query_embd)
  retrieved_id = np.argmax(similarities)
  s = data_collection[retrieved_id]
  index_of_hash = s.index('#')
  # 提取#前面的数字
  number_before_hash = ''
  i = index_of_hash - 1
  while s[i].isdigit():
    number_before_hash = s[i] + number_before_hash
    i -= 1
  return number_before_hash
```

```python
# 开始产出
review_ids=[]
description_ids=[]
for index_x, row_x in df_description.iterrows():
  description = row_x["description"]
  description_index = row_x["index"]
  ans = getRelateTrue(description)
  review_ids.append(ans)
  description_ids.append(description_index)

output_file_path = "/content/drive/MyDrive/movies_analysis/movies_analysis_description.xlsx"

data = {
    'review_id': review_ids,
    'description_id': description_ids
}
df = pd.DataFrame(data)

df.to_excel(output_file_path, index=False)
# 看看文件
df_raw2 = pd.read_excel(output_file_path)
df_raw2.head()
```
